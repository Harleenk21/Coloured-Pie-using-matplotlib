```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(12,3))

plt.pie([1],colors="k",radius=2.05)
labels=np.ones(20)
clr=['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
plt.pie(labels,colors=clr,radius=2.0)

plt.pie([1],colors='g',radius=1.8)
plt.pie([1],colors='y',radius=1.5)
plt.pie([1],colors='c',radius=1.3)
plt.pie([1],colors='b',radius=1.0)
plt.pie([1],colors='m',radius=0.8)

plt.pie([1],colors='b',radius=0.31)
plt.pie(labels,colors=clr,radius=0.3)

plt.pie([1],colors='w',radius=0.2)
plt.pie([1],colors='k',radius=0.1)
plt.show()
```
